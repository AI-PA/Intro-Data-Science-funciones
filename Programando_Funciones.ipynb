```python
import matplotlib.pyplot as plt 
import numpy as np
```

# Funciones Algebraicas

## Funcion lineal

```python
# Definiendo Constantes y tamaño del gráfico
N=100
m=10
b=5

# Función a evaluar
def f(x):
    return m*x+b 

# Asignación de valores
x=np.linspace(-10,10,num=N)
y = f(x)

# Graficando resultados
fig,ax=plt.subplots()
ax.plot(x,y)
ax.grid()
plt.show()
```
